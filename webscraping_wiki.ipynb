```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'

page = requests.get(url)

soup = bs(page.text, 'html')
```

```python
# print(soup)
```

```python
table = soup.find_all('table')[0]
# print(table)
```

```python
trs = table.find_all('tr')[0]
titles = trs.find_all('th')
```

```python
header = [title.text.strip() for title in titles]
header[6] = 'Headquarters'
header = header[:7]
print(header)
```

```python
import pandas as pd

df = pd.DataFrame(columns = header)
df
```

```python
rows = table.find_all('tr')[2:]

for row in rows:

  th = row.find('th')
  rank = th.text.strip()
  td = row.find_all('td')
  data = [dt.text.strip() for dt in td]
  data.insert(0,rank)
  data = data[:7]
  # print(data)
  length = len(df)
  df.loc[length] = data

df = df.drop_duplicates()
df
```

```python
df.to_csv(r'companies.csv', index = False)
```
